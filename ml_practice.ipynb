# One Hot Encoder

```python
# one hot encoder: used when working with categorical data
# transform categorical
# nominal data: cate data without numeric difference (colors)
from sklearn.preprocessing import OneHotEncoder
```

```python
d = {'sales': [100000,222000,1000000,522000,111111,222222,1111111,20000,75000,90000,1000000,10000], 'city': ['Tampa','Tampa','Orlando','Jacksonville','Miami','Jacksonville','Miami','Miami','Orlando','Orlando','Orlando','Orlando'], 'size': ['Small', 'Medium','Large','Large','Small','Medium','Large','Small','Medium','Medium','Medium','Small',]}
```

```python
df = pd.DataFrame(data=d)
```

```python
df.head()
```

```python
# choose data that doesnt have numeric difference
# city, doesnt have numeric difference
df['city'].unique()
```

```python
# one hot encoder expands the table with the cities, change cities within pandas df
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False).set_output(transform='pandas')
```

```python
ohetransform = ohe.fit_transform(df[['city']])
```

```python
ohetransform
# first data is tampa, 2nd is tampa, 3rd is orlando,
# abc order automatically
```

```python
df.head()
```

```python
ohetransform.head()
```

```python
# merge df with ohetransform
df = pd.concat([df, ohetransform], axis=1).drop(columns=['city'])
```

```python
df.head(10)
```

# Ordinal Encoder

transforms ordinal data (inherit ranking like size) into numerical data like 0 for small, 1 for med, 2 for large

```python
d = {'sales': [100000,222000,1000000,522000,111111,222222,1111111,20000,75000,90000,1000000,10000], 'city': ['Tampa','Tampa','Orlando','Jacksonville','Miami','Jacksonville','Miami','Miami','Orlando','Orlando','Orlando','Orlando'], 'size': ['Small', 'Medium','Large','Large','Small','Medium','Large','Small','Medium','Medium','Medium','Small',]}
df = pd.DataFrame(data=d)
```

```python
df.head()
```

```python
# sales for difference stores and cities
# categorical data, ordinal(inherit ranking): size, nominal(no inherit ranking): city

#specify order
df['size'].unique()
```

```python
sizes = ['Small', 'Medium', 'Large']
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
# allows small for first, med for second, large for third. dont want to be in abc order
enc = OrdinalEncoder(categories=[sizes])
```

```python
enc.fit_transform(df[['size']])
```

```python
df.head()
```

```python
df['size'] = enc.fit_transform(df[['size']])
```

```python
df.head()
```

# Simple Imputer
- fill in null values in pandas df
- most ml models don't allow null values so simple imputer is a good way to fix that
- mode, mean, median

```python
import numpy as np
```

```python
miles = pd.DataFrame({'farthest_run_mi': [50,62,np.nan,100,26,13,31,50]})
```

```python
miles
```

```python
miles.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp_mean = SimpleImputer(strategy='mean') # most commonly used, depending on where data lies
```

```python
imp_mean.fit_transform(miles)
```

median

```python
imp_median = SimpleImputer(strategy='median')
```

```python
imp_median.fit_transform(miles)
```

Mode
- can also be used for categorical data

```python
imp_mode = SimpleImputer(strategy='most_frequent')
imp_mode.fit_transform(miles)
```

constant

```python
imp_constant = SimpleImputer(strategy='constant', fill_value=13)
imp_constant.fit_transform(miles)
```

Categorical

```python
names = pd.DataFrame({'names': ['ryan', 'nolan', 'honus', 'wagner', np.nan, 'ruth']})
```

```python
names
```

```python
imp_constant_category = SimpleImputer(strategy='constant', fill_value='babe')
```

```python
imp_constant_category.fit_transform(names)
```

show imputed

```python
# add brand new column of 0s and 1s of what values were imputed
imp_mean_marked = SimpleImputer(strategy='mean', add_indicator=True)
```

```python
imp_mean_marked.fit_transform(miles)
```

More complicated example

```python
df = pd.DataFrame({'Name': ['Ryan', 'Nolan', 'Walter', 'Honus', 'Christy', np.nan, 'Napolean', 'Tris'],
                   'farthest_run_mi': [50,62,np.nan,100,26,13,31,50]})
```

# Column Transformer

```python
d = {'sales': [100000,222000,1000000,522000,111111,222222,1111111,20000,75000,90000,1000000,10000], 'city': ['Tampa','Tampa','Orlando','Jacksonville','Miami','Jacksonville','Miami','Miami','Orlando','Orlando','Orlando','Orlando'], 'size': ['Small', 'Medium','Large','Large','Small','Medium','Large','Small','Medium','Medium','Medium','Small',]}
df = pd.DataFrame(data=d)
```

```python
df
```

```python
# one hot encoder
# ordinal encoder
ohe = OneHotEncoder(sparse_output=False)
ode = OrdinalEncoder()
```

```python
# import make column transformer

ct = make_column_transformer(
    # one hot encoder, column to implement
    (ohe, ['city']),
    # ordinal encoder,
    (ode, ['size']),
    # what happens with the remaining cols, passthrough if you want to keep cols
    remainder = 'passthrough'
)
```

```python
ct.set_output(transform='pandas')
```

```python
df_pandas = ct.fit_transform(df)
```

```python
df_pandas
```

```python
# drop
ct2 = make_column_transformer(
    # one hot encoder, column to implement
    (ohe, [1]),
    # ordinal encoder,
    (ode, [2]),
    # what happens with the remaining cols, passthrough
    remainder = 'drop'
)
```

```python
ct2.set_output(transform='pandas')
```

```python
df_pandas2 = ct2.fit_transform(df)
```

```python
df_pandas2
```

```python
# drop certain cols
ct3 = make_column_transformer(
    # one hot encoder, column to implement
    (ohe, ['city']),
    # ordinal encoder,
    ('passthrough', ['size']),
    # what happens with the remaining cols, passthrough
    remainder = 'drop'
)
```

```python
ct3.set_output(transform='pandas')
```

```python
df_pandas3 = ct3.fit_transform(df)
```

```python
df_pandas3
```

# k nearest neighbors

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# analyze results
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

Code the model

```python
# code out the model
df = pd.read_csv('500hits.csv', encoding='latin-1')
```

```python
df.head()
```

cleanup, cant have player names rather have only numerical values, drop unneccessary data columns

```python
df = df.drop(columns=['PLAYER', 'CS'])
```

```python
df.head()
```

```python
# separate into x features and y outcome
# what causes for someone to be in the hall of fame?
# y = hall of fame
# x =

X = df.iloc[:,0:13]
```

```python
y = df.iloc[:,13]
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=11, test_size=0.2)
```

```python
# add in a scaler
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
# fit data into scaler
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
# add in knn
knn = KNeighborsClassifier(n_neighbors=8)
```

```python
# fit data into model
knn.fit(X_train, y_train)
```

```python
# make some predictions
y_pred = knn.predict(X_test)
```

```python
print(y_pred)
```

```python
knn.score(X_test, y_test)
```

```python
# reporting: confusion matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
print(cm)

# top left: true positive
# bottom right: true negative
# top right: false positive
# bottom left: false negative
```

```python
cr = classification_report(y_test, y_pred)
```

```python
print(cr)

# precision
# 55/59 = 0.93

# recall
# true pos/(trpos+false pos), 55/(55+12) = 0.82

# f1-score
# 2*0.93* = 1.525
# 0.93+0.83 = 1.75
# 1.525/1.75 = 0.87

# support
# 55 + 12 = 67
# 4 + 22 = 26

# also took out outliers
# 0.93 + 0.65 / 2 = 0.79 = macro avg precision
```

```python
print(knn.n_samples_fit_)
# 372+93 = 465
```
