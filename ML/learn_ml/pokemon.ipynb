```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("pokemon.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
# look at the number of non-legendary and legendary pokemon
sns.countplot(x='is_legendary', data=df)
```

```python
# clean dataset with categorical and missing values

df_cleaned = df.select_dtypes(exclude=['object']) # exclude categorical values/objects
df_cleaned = df_cleaned.dropna() # dropped cols with NaN values

print(df_cleaned.shape)
df_cleaned.head()
```

```python
sns.countplot(x="is_legendary", data=df_cleaned)
```
